```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
data = pd.read_csv("../raw-data/Diet-Compositions/Diet compositions by commodity categories - FAO (2017).csv")
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.describe(include = "all")
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)] # alcholic beverages only have null values
df["Alcoholic Beverages (FAO (2017)) (kilocalories per person per day)"] = df["Alcoholic Beverages (FAO (2017)) (kilocalories per person per day)"].fillna(0)
```

```python
df.isnull().sum()
```

```python
food_categories = {
    "Cereals and Grains (FAO (2017)) (kilocalories per person per day)": "Cereals & Grains",
    "Pulses (FAO (2017)) (kilocalories per person per day)": "Pulses",
    "Starchy Roots (FAO (2017)) (kilocalories per person per day)": "Starchy Roots",
    "Sugar (FAO (2017)) (kilocalories per person per day)": "Sugar",
    "Oils & Fats (FAO (2017)) (kilocalories per person per day)": "Oils & Fats",
    "Meat (FAO (2017)) (kilocalories per person per day)": "Meat",
    "Dairy & Eggs (FAO (2017)) (kilocalories per person per day)": "Dairy & Eggs",
    "Fruit and Vegetables (FAO (2017)) (kilocalories per person per day)": "Fruits & Vegetables",
    "Other (FAO (2017)) (kilocalories per person per day)": "Other",
    "Alcoholic Beverages (FAO (2017)) (kilocalories per person per day)": "Alcoholic Beverages"
}
```

```python
avg_data = df.groupby('Year')[list(food_categories.keys())].mean().reset_index()

fig = go.Figure()

colors = ['#e74c3c', '#3498db', '#2ecc71', '#f39c12', '#9b59b6',
          '#1abc9c', '#e67e22', '#34495e', '#95a5a6', '#d35400']

color_idx = 0
for full_name, short_name in food_categories.items():
    fig.add_trace(go.Scatter(
        x=avg_data['Year'],
        y=avg_data[full_name],
        name=short_name,
        mode='lines',
        stackgroup='one',
        fillcolor=colors[color_idx],
        line=dict(width=0.5, color=colors[color_idx]),
        hovertemplate='<b>%{fullData.name}</b><br>Year: %{x}<br>Avg: %{y:.1f} kcal/day<extra></extra>'
    ))
    color_idx += 1

fig.update_layout(
    title='Global Food Consumption Composition',
    xaxis_title='Year',
    yaxis_title='Total Kilocalories per person per day',
    width=1500,
    height=700,
    hovermode='x unified',
    plot_bgcolor='white',
    xaxis=dict(dtick=10, tickformat='d')
)

fig.show()
```

```python
latest_year = 2013
df_latest = df[df['Year'] == latest_year].copy()

fig = make_subplots(
    rows=2, cols=5,
    subplot_titles=list(food_categories.values()),
    vertical_spacing=0.15,
    horizontal_spacing=0.06
)

row_col_pairs = [(1,1), (1,2), (1,3), (1,4), (1,5),
                 (2,1), (2,2), (2,3), (2,4), (2,5)]

category_colors = {
    'Cereals & Grains': '#f39c12',
    'Pulses': '#27ae60',
    'Starchy Roots': '#d35400',
    'Sugar': '#e74c3c',
    'Oils & Fats': '#f1c40f',
    'Meat': '#c0392b',
    'Dairy & Eggs': '#3498db',
    'Fruits & Vegetables': '#2ecc71',
    'Other': '#95a5a6',
    'Alcoholic Beverages': '#9b59b6'
}

for idx, (full_name, short_name) in enumerate(food_categories.items()):
    row, col = row_col_pairs[idx]

    df_cat = df_latest[['Entity', full_name]].copy()
    df_cat = df_cat.sort_values(full_name, ascending=False).head(8)

    fig.add_trace(
        go.Bar(
            x=df_cat['Entity'],
            y=df_cat[full_name],
            marker=dict(
                color=category_colors[short_name],
                line=dict(color='white', width=1.5)
            ),
            showlegend=False,
            text=df_cat[full_name].round(0),
            textposition='outside',
            textfont=dict(size=9),
            hovertemplate='<b>%{x}</b><br>' + short_name + ': %{y:.1f} kcal/day<extra></extra>'
        ),
        row=row,
        col=col
    )

fig.update_layout(
    title={
        'text': f'Top 8 Countries by Food Category ({latest_year})<br>',
        'x': 0.5,
        'xanchor': 'center',
        'font': {'size': 22, 'color': '#2c3e50'}
    },
    height=950,
    width=2100,
    showlegend=False,
    plot_bgcolor='white',
    paper_bgcolor='#f8f9fa'
)

fig.update_xaxes(
    showgrid=False,
    tickangle=-45,
    tickfont=dict(size=9),
    showline=True,
    linewidth=1,
    linecolor='#95a5a6'
)
fig.update_yaxes(
    showgrid=True,
    gridcolor='#e0e0e0',
    tickfont=dict(size=9),
    showline=True,
    linewidth=1,
    linecolor='#95a5a6'
)

fig.show()
```

```python
calorie_col = "Cereals and Grains (FAO (2017)) (kilocalories per person per day)"

fig = go.Figure()

countries = sorted(df["Entity"].unique())

for country in countries:
    country_data = df[df["Entity"] == country].sort_values("Year")

    fig.add_trace(go.Scatter(
        x=country_data["Year"],
        y=country_data[calorie_col],
        mode='lines+markers',
        name=country,
        line=dict(width=2.5),
        marker=dict(size=5),
        visible='legendonly',
        hovertemplate='<b>%{fullData.name}</b><br>' +
                      'Year: %{x}<br>' +
                      'Consumption: %{y:.1f} kcal/day<br>' +
                      '<extra></extra>'
    ))

fig.update_layout(
    title={
        'text': "Cereals and Grains Consumption by Country (1961-2020)<br>" +
                "<sub>Click country names to toggle | Double-click to isolate | Shift+click for multiple</sub>",
        'x': 0.5,
        'xanchor': 'center',
        'font': {'size': 18, 'color': '#2c3e50'}
    },
    xaxis_title="Year",
    yaxis_title="Kilocalories per person per day",
    width=1500,
    height=850,
    hovermode='closest',
    legend=dict(
        title=dict(text='<b>Countries</b><br>(Click to show)', font=dict(size=12)),
        yanchor="top",
        y=0.98,
        xanchor="left",
        x=1.01,
        font=dict(size=9),
        itemsizing='constant',
        bgcolor='rgba(255,255,255,0.9)',
        bordercolor='Gray',
        borderwidth=1
    ),
    font=dict(size=12, family="Arial"),
    plot_bgcolor='white',
    paper_bgcolor='#f8f9fa'
)

fig.update_xaxes(
    showgrid=True,
    gridwidth=1,
    gridcolor='#e0e0e0',
    showline=True,
    linewidth=2,
    linecolor='black'
)
fig.update_yaxes(
    showgrid=True,
    gridwidth=1,
    gridcolor='#e0e0e0',
    showline=True,
    linewidth=2,
    linecolor='black'
)

fig.show()
```

```python
df_2013 = df[df['Year'] == 2013]
avg_2013 = df_2013[list(food_categories.keys())].mean()

fig = go.Figure(data=[go.Pie(
    labels=list(food_categories.values()),
    values=avg_2013.values,
    hole=0.4,
    marker=dict(
        colors=['#f39c12', '#27ae60', '#d35400', '#e74c3c', '#f1c40f',
                '#c0392b', '#3498db', '#2ecc71', '#95a5a6', '#9b59b6'],
        line=dict(color='white', width=2)
    ),
    textinfo='label+percent',
    textfont=dict(size=12),
    hovertemplate='<b>%{label}</b><br>%{value:.1f} kcal/day<br>%{percent}<extra></extra>'
)])

fig.update_layout(
    title={
        'text': 'Global Food Consumption Composition (2013)<br><sub>Average Daily Calorie Intake by Source</sub>',
        'x': 0.5,
        'xanchor': 'center',
        'font': {'size': 22, 'color': '#2c3e50'}
    },
    width=1000,
    height=700,
    showlegend=True,
    legend=dict(
        orientation="v",
        yanchor="middle",
        y=0.5,
        xanchor="left",
        x=1.05,
        font=dict(size=11)
    ),
    paper_bgcolor='#f8f9fa'
)

fig.show()
```

```python
selected_countries = ['United States', 'China', 'India', 'Brazil']

df_2013 = df[df['Year'] == 2013]

fig = make_subplots(
    rows=2, cols=2,
    specs=[[{"type": "pie"}, {"type": "pie"}],
           [{"type": "pie"}, {"type": "pie"}]],
    subplot_titles=selected_countries
)

colors = ['#f39c12', '#27ae60', '#d35400', '#e74c3c', '#f1c40f',
          '#c0392b', '#3498db', '#2ecc71', '#95a5a6', '#9b59b6']

positions = [(1,1), (1,2), (2,1), (2,2)]

for idx, country in enumerate(selected_countries):
    row, col = positions[idx]
    country_data = df_2013[df_2013['Entity'] == country][list(food_categories.keys())].iloc[0]

    fig.add_trace(go.Pie(
        labels=list(food_categories.values()),
        values=country_data.values,
        hole=0.3,
        marker=dict(colors=colors, line=dict(color='white', width=2)),
        textinfo='percent',
        textfont=dict(size=10),
        hovertemplate='<b>%{label}</b><br>%{value:.1f} kcal/day<br>%{percent}<extra></extra>'
    ), row=row, col=col)

fig.update_layout(
    title={
        'text': 'Food Consumption Composition by Country (2013)',
        'x': 0.5,
        'xanchor': 'center',
        'font': {'size': 22, 'color': '#2c3e50'}
    },
    width=1400,
    height=1000,
    showlegend=True,
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=-0.15,
        xanchor="center",
        x=0.5,
        font=dict(size=11)
    ),
    paper_bgcolor='#f8f9fa'
)

fig.show()
```
