```python
gdp_df = df1[df1['Series Name'] == 'GDP per capita (constant 2015 US$)'].copy()
print("Filtered df1 (gdp_df) head:")
print(gdp_df.head())
```

```python
year_cols = [col for col in gdp_df.columns if 'YR' in col]
gdp_melted = gdp_df.melt(id_vars=['Country Name', 'Country Code', 'Series Name', 'Series Code'], 
                       value_vars=year_cols, 
                       var_name='Year', 
                       value_name='Per_Capita_Income')
print("Melted gdp_df head:")
print(gdp_melted.head())
```

```python
gdp_melted['Year'] = gdp_melted['Year'].str.extract(r'(\d{4})').astype(int)
print("Cleaned 'Year' column in gdp_melted head:")
print(gdp_melted.head())
```

```python
gdp_melted['Per_Capita_Income'] = pd.to_numeric(gdp_melted['Per_Capita_Income'], errors='coerce')
print("Cleaned 'Per_Capita_Income' column in gdp_melted head:")
print(gdp_melted.head())
```

```python
gdp_melted.rename(columns={'Country Name': 'Entity'}, inplace=True)
print("Renamed 'Country Name' to 'Entity' in gdp_melted head:")
print(gdp_melted.head())
```

```python
merged_df = pd.merge(df, gdp_melted[['Entity', 'Year', 'Per_Capita_Income']], on=['Entity', 'Year'], how='inner')
print("Merged DataFrame head:")
print(merged_df.head())
print("\nMerged DataFrame info:")
merged_df.info()
```

```python
import matplotlib.pyplot as plt

# Identify diet category columns
diet_columns = [
    'Cereals and Grains (FAO (2017)) (kilocalories per person per day)',
    'Pulses (FAO (2017)) (kilocalories per person per day)',
    'Starchy Roots (FAO (2017)) (kilocalories per person per day)',
    'Sugar (FAO (2017)) (kilocalories per person per day)',
    'Oils & Fats (FAO (2017)) (kilocalories per person per day)',
    'Meat (FAO (2017)) (kilocalories per person per day)',
    'Dairy & Eggs (FAO (2017)) (kilocalories per person per day)',
    'Fruit and Vegetables (FAO (2017)) (kilocalories per person per day)',
    'Other (FAO (2017)) (kilocalories per person per day)',
    'Alcoholic Beverages (FAO (2017)) (kilocalories per person per day)'
]

# Select a few diverse countries for visualization
selected_countries = ['United States', 'China', 'India', 'France', 'Brazil']

# Create plots for each selected country
for country in selected_countries:
    country_df = merged_df[merged_df['Entity'] == country].copy()

    if not country_df.empty:
        fig, axes = plt.subplots(2, 1, figsize=(12, 10), sharex=True)
        fig.suptitle(f'Diet Composition and Per Capita Income Trends in {country}', fontsize=16)

        # Plot diet categories
        for col in diet_columns:
            axes[0].plot(country_df['Year'], country_df[col], label=col.split('(')[0].strip())
        axes[0].set_ylabel('Kilocalories per person per day')
        axes[0].set_title('Diet Category Trends')
        axes[0].legend(loc='upper left', bbox_to_anchor=(1, 1))
        axes[0].grid(True)

        # Plot per capita income
        axes[1].plot(country_df['Year'], country_df['Per_Capita_Income'], label='Per Capita Income', color='purple')
        axes[1].set_xlabel('Year')
        axes[1].set_ylabel('Per Capita Income (USD)')
        axes[1].set_title('Per Capita Income Trend')
        axes[1].grid(True)
        axes[1].legend(loc='upper left', bbox_to_anchor=(1, 1))

        plt.tight_layout(rect=[0, 0.03, 0.9, 0.95])
        plt.show()
    else:
        print(f"No data found for {country}")
```
