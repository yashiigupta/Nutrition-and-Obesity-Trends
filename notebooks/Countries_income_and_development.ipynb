```python
import pandas as pd
```

```python
countries_gdp = pd.read_csv('../raw-data/Income-GDP/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_216063.csv', skiprows=4)
countries_income_group = pd.read_csv('../raw-data/Income-GDP/Metadata_Country_API_NY.GDP.MKTP.CD_DS2_en_csv_v2_216063.csv')
```

```python
gdp_filtered = countries_gdp[['Country Name', 'Country Code', '2000', '2022' ]]
gdp_filtered = gdp_filtered.rename(columns={'2024': 'GDP_2024'})
gdp_filtered.isnull().sum()
```

```python
gdp_filtered.duplicated().sum()
```

```python
income_group_filtered = countries_income_group[['Country Code', 'IncomeGroup', 'SpecialNotes']]
income_group_filtered.isnull().sum()
```

```python
merged_df = pd.merge(gdp_filtered, income_group_filtered, on='Country Code', how='inner')
merged_df.isnull().sum()
```

```python
merged_df['GDP_2024_Trillions'] = merged_df['GDP_2024'] / 1e12
merged_df = merged_df.drop(columns=['GDP_2024'])
```

```python
null_gdp_income_df = merged_df[merged_df['GDP_2024_Trillions'].isnull() | merged_df['IncomeGroup'].isnull()]
null_gdp_income_df
```

```python
merged_df.to_csv('../data/merged_gdp_income_data.csv', index=False)
print("merged")
```
