### **What overall pattern emerges between rising GDP and obesity rates in countries?**

```python
import pandas as pd
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import math
```

```python
df_gdp = pd.read_csv('../data/clean_gdp_data.csv')
df_obesity = pd.read_csv("../data/clean_obesity_data.csv")
```

```python
display(df_gdp)
display(df_obesity)
```

```python
country_name_mapping = {
    'Bahamas, The': 'Bahamas',
    'Bolivia': 'Bolivia (Plurinational State of)',
    'Congo, Rep.': 'Congo',
    'Congo, Dem. Rep.': 'Democratic Republic of the Congo',
    "Cote d'Ivoire": "Côte d'Ivoire",
    'Egypt, Arab Rep.': 'Egypt',
    'Gambia, The': 'Gambia',
    'Iran, Islamic Rep.': 'Iran (Islamic Republic of)',
    'Korea, Rep.': 'Republic of Korea',
    'Kyrgyz Republic': 'Kyrgyzstan',
    'Lao PDR': "Lao People's Democratic Republic",
    'Micronesia, Fed. Sts.': 'Micronesia (Federated States of)',
    'Netherlands': 'Netherlands (Kingdom of the)',
    'Slovak Republic': 'Slovakia',
    'Somalia, Fed. Rep.': 'Somalia',
    'Tanzania': 'United Republic of Tanzania',
    'Turkiye': 'Türkiye',
    'United Kingdom': 'United Kingdom of Great Britain and Northern Ireland',
    'United States': 'United States of America',
    'Venezuela, RB': 'Venezuela (Bolivarian Republic of)',
    'Yemen, Rep.': 'Yemen',
    'West Bank and Gaza': 'Palestine'
}

# Apply mapping to GDP dataset
df_gdp['Country Name'] = df_gdp['Country Name'].replace(country_name_mapping)
```

```python
df_obesity_melted = pd.melt(
    df_obesity,
    id_vars=['Area'],
    var_name='Year',
    value_name='Obesity_Rate'
).rename(columns={'Area': 'Country Name'})

df_gdp_melted = pd.melt(
    df_gdp,
    id_vars=['Country Name', 'Country Code'],
    var_name='Year',
    value_name='GDP_Value'
)

df_obesity_melted['Year'] = pd.to_numeric(df_obesity_melted['Year'])
df_gdp_melted['Year'] = pd.to_numeric(df_gdp_melted['Year'])
```

```python
gdp_countries = set(df_gdp_melted['Country Name'].unique())
obesity_countries = set(df_obesity_melted['Country Name'].unique())

common_countries = gdp_countries.intersection(obesity_countries)

df_gdp_melted = df_gdp_melted[df_gdp_melted['Country Name'].isin(common_countries)]
df_obesity_melted = df_obesity_melted[df_obesity_melted['Country Name'].isin(common_countries)]
```

```python
df_combined = pd.merge(
    df_gdp_melted,
    df_obesity_melted,
    on=['Country Name', 'Year'],
    how='inner'
)

print("Final Combined Data:")
print(df_combined.head())
print(df_combined.info())
```

```python

selected_countries = [
    'Australia',
    'Bangladesh',
    'Bhutan',
    'Brazil',
    'Canada',
    'China',
    'France',
    'Germany',
    'Japan',
    'Nepal',
    'Pakistan',
    'Russian Federation',
    'South Africa',
    'Sri Lanka',
    'Switzerland',
    'Türkiye',
    'United Kingdom of Great Britain and Northern Ireland',
    'United States of America'
]

cols = 3
n = len(selected_countries)
rows = math.ceil(n / cols)

fig = make_subplots(
    rows=rows,
    cols=cols,
    subplot_titles=selected_countries,
    specs=[[{"secondary_y": True} for _ in range(cols)] for _ in range(rows)]
)

for i, country in enumerate(selected_countries):
    r = (i // cols) + 1
    c = (i % cols) + 1
    country_data = df_combined[df_combined['Country Name'] == country].sort_values(by='Year')
    
    # For the first subplot → show legend
    show_legend = (i == 0)

    # GDP (Blue)
    fig.add_trace(
        go.Scatter(
            x=country_data['Year'],
            y=country_data['GDP_Value'],
            name="GDP Value",
            line=dict(color='blue'),
            legendgroup="GDP",
            showlegend=show_legend
        ),
        row=r, col=c, secondary_y=False
    )
    
    # Obesity (Red)
    fig.add_trace(
        go.Scatter(
            x=country_data['Year'],
            y=country_data['Obesity_Rate'],
            name="Obesity Rate",
            line=dict(color='red'),
            legendgroup="Obesity",
            showlegend=show_legend
        ),
        row=r, col=c, secondary_y=True
    )

    fig.update_xaxes(title_text="Year", row=r, col=c)
    fig.update_yaxes(title_text="GDP (Billions USD)", row=r, col=c, secondary_y=False)
    fig.update_yaxes(title_text="Obesity (%)", row=r, col=c, secondary_y=True)

fig.update_layout(
    height=300 * rows,
    width=1400,
    title_text="GDP vs Obesity Rate for All Countries",
    hovermode="x unified"
)

fig.show()
```

### Key Observations from Country Trends

#### General GDP Growth
- Most countries (e.g., U.S., India, Germany, China, U.K.) show a clear upward GDP trend, reflecting steady economic expansion.

#### Rising Obesity Rates
- Obesity rates also rise across nearly all countries, indicating a growing global health concern linked to modern lifestyles.

#### Apparent Co-Movement
- In many cases, GDP and obesity increase together, suggesting a general positive association—though not direct causation.

#### Differences in Pace
- The rate of increase varies by country:
    - Some show fast GDP growth and sharp obesity rises.
    - Others display slower, gradual changes in both.
- These variations reflect differences in culture, policy, and development levels.

### Comparative Analysis of Outlier Countries

#### France
- **GDP**: Generally rising with some normal fluctuations.
- **Obesity**: Increasing, but at a slower and more controlled rate, likely influenced by Mediterranean dietary habits and lifestyle.

#### Japan
- **GDP**: High and stable, with modest growth typical of a mature economy.
- **Obesity**: Very low and slow-rising, making Japan a clear outlier due to strong cultural dietary patterns and active lifestyle norms.
