```python
df = df.drop(columns=['Domain Code', 'Domain', 'Area Code (M49)', 'Element Code', 'Element', 'Item Code', 'Item', 'Year Code', 'Unit', 'Flag', 'Flag Description', 'Note'])
df.head()
```

```python
df_pivot = df.pivot_table(index='Area', columns='Year', values='Value')
df_pivot.head()
```

```python
df_pivot = df_pivot.reset_index()
df_pivot.head()
```

```python
df_pivot.to_csv('../data/clean_obesity_data.csv', index=False)
print("df_pivot saved as 'clean_obesity_data.csv'")
```
